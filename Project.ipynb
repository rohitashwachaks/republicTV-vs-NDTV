```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
texas_capitol = pd.read_csv('CapitolBuilding.csv', header=None)
```

```python
texas_capitol.head()
```

```python
texas_capitol = texas_capitol[0].apply(word_tokenize)
```

```python
texas_capitol
```

```python
stop_words = set(stopwords.words('english'))
stop_words.add('.')
stop_words.add('!')
stop_words.add('..')
stop_words.add('...')
stop_words.add(',')
```

```python
## Tokenizing, removing stop words, finding frequency
tok = []
for i in range(0,len(texas_capitol)):
    for j in texas_capitol[i]:
        if (j.lower() not in stop_words):
            tok.append(j)    
tok
```

```python
fdist = FreqDist(tok)
```

```python
fdist
```

```python
tags = nltk.pos_tag(tok)
```

```python
print('parts of speech ', tags )
```

```python
adj_tags = [t for t in tags if t[1] == "JJ"]
adj_tags
```
